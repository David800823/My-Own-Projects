```python
# We import Fred from the fredapi. This helps us get data straight from the 'https://fred.stlouisfed.org/' website
from fredapi import Fred

# Importing Libraries
import pandas as pd
import numpy as np
```

```python
#List of the Series ID

series_data = {
"CPI" : "CPIAUCSL",  #CPI US City Average
"Core_CPI" : "CPILFESL", #Core CPI US City Average
"Chicken_Breat" : "APU0000FF1101", #Average Price for Chicken Breast, Boneless per pound US City Average
"Eggs" : "APU0000708111", #Average Price Eggs, Grade A, Large (Cost per Dozen)
"Milk" : "APU0000709112", #Average Price: Milk, Fresh, Whole, Cost per gallon
"White_Bread" : "APU0000702111", # Average price of White Bread per pound 
"Gasoline_All_Types" : "CUSR0000SETB01", # All type of gsoline price
"Used_Car_and_Truck" : "CUSR0000SETA02", # Used Car Price
"New_Car" : "CUSR0000SETA01", #New Car Prices
"Federal_Funds_Rate" : "FEDFUNDS"

}


```

```python
for series_name, series_id in series_data.items():
  series_data[series_name] = fred.get_series(series_id = series_id, observation_start = '01/01/2006')
```

```python
CPI_all_values = pd.DataFrame.from_dict(series_data)
```
